```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.fft import fft, ifft
```

```python
def make_rect_like(points:np.ndarray):
    x= np.zeros(len(points)*2)
    y = np.zeros(len(points)*2)
    for i in range(len(points) - 1):
        x[i*2+0] = x[i*2+1] = i
        y[i*2+0] = points[i]
        y[i*2+1] = points[i+1]

    x[-1] = len(points) - 2
    x[-2] = len(points) - 2
    y[-1] = points[-1]
    return x, y
```

```python
sampling_rate = 130
ts = 1/sampling_rate
start = 0
end = 3
t = np.arange(start, end, ts)

f1 = 2
x1 = np.sin(2 * np.pi * f1 * t) # real signal

f2 = 30
noise = 0.2*np.sin(2 * np.pi * f2 * t) # noise signal

fig, ax = plt.subplots(nrows= 3, ncols= 1)
fig.set_figwidth(12)
fig.set_figheight(5)
fig.tight_layout(h_pad=2)
ax[0].plot(t, x1)
ax[0].set_title('Pure Signal')
ax[1].plot(t, noise)
ax[1].set_title('Noise')
ax[2].plot(t, x1 + noise)
ax[2].set_title('Signal With Noise')
```

```python
Y = fft(x1 + noise)
N = len(Y)
Y_mag = np.abs(Y) / N
n = np.arange(N)
T = N/sampling_rate
freq = n/T  
print(f'{N=}, {len(freq)=}, {T=}')

f_plot = freq[0:int(N/2+1)]
X_mag_plot = 2 * Y_mag[0:int(N/2+1)]
X_mag_plot[0] = X_mag_plot[0] / 2

fig, ax = plt.subplots(nrows=4, ncols=1)
fig.set_figwidth(12)
fig.set_figheight(5)
fig.tight_layout(h_pad=2)

to_freq = lambda i: i * sampling_rate / N

frequencies = to_freq(n)
half = int(len(frequencies) * 0.5)

ax[0].plot(frequencies[:half], abs(Y)[:half])

m = [0 if to_freq(i) > 10 and to_freq(i) < to_freq(len(Y) - 1) - 10 else 1 for i in range(len(Y))]
m = np.array(m)
ax[1].plot(frequencies[:half], abs(m)[:half])

YL = Y * m

ax[2].plot(frequencies[:half], abs(YL)[:half])

x_ = ifft(YL) 

ax[3].plot(t, x_)
```

```python
def low_pass_filter(x:np.ndarray, max_freq) -> np.ndarray:
    X = fft(x)
    N = len(x)
    
    to_freq = lambda i: i * sampling_rate / N

    m = [0 if to_freq(i) > max_freq and to_freq(i) < to_freq(len(X) - 1) - max_freq else 1 for i in range(len(X))]
    m = np.array(m) 

    return ifft(X * m) 
```

```python
x = np.sin(2 * np.pi * 2 * t) + 0.4 * np.sin(2 * np.pi * 30 * t)
x_ = low_pass_filter(x, 10)

fig, axis = plt.subplots(nrows= 2, ncols= 1)
fig.tight_layout(h_pad=3)
fig.set_figwidth(12)

axis[0].plot(t, x)
axis[0].set_title('Signal With Noise')

axis[1].plot(t, x_)
axis[1].set_title('Low Pass Signal')
```

```python
def delta_modulation(x:np.ndarray, delta:float=0.1) -> np.ndarray:
    mx = np.zeros(x.shape)
    va = np.zeros(x.shape)
    v = x[0]
    for i in range(len(x)):
        difference = x[i] - v
        v += delta if difference >= 0 else -delta
        mx[i] = 1 if difference >= 0 else 0

        va[i] = v
    return mx, va
```

```python
x = np.sin(2 * np.pi * 2 * t)
x = np.ones(t.shape)

mx, va = delta_modulation(x)
fig, axis = plt.subplots(nrows=3, ncols=1)
fig.set_figwidth(15)
axis[0].plot(x)

vax, vay = make_rect_like(va)
axis[1].plot(vax[:200], vay[:200])
mxx, mxy = make_rect_like(mx)
axis[2].plot(mxx[:200], mxy[:200])
```

```python
def delta_demodulation(mx:np.ndarray, delta:float=0.1) -> np.ndarray:
    x = np.zeros(mx.shape)
    for i in range(1, len(x)):
        dir = mx[i] * 2 - 1
        x[i] = x[i - 1] + delta * dir 
    return x
```

```python
x = np.sin(2 * np.pi * 2 * t)


delta = 0.1

mx, va = delta_modulation(x, delta)
dmx = delta_demodulation(mx, delta)

fig, axis = plt.subplots(nrows=5, ncols=1)
fig.set_figwidth(12)
fig.set_figheight(8)
fig.tight_layout(h_pad=3)

axis[0].plot(x)
axis[0].set_title('Oignal Signal')

axis[1].set_title('Difference Modulation Signal')
va_x, va_y = make_rect_like(va)
axis[1].plot(va_x, va_y)

axis[2].set_title('Delta Modulation Signal')
mx_x, mx_y = make_rect_like(mx[:150])
axis[2].plot(mx_x, mx_y)

axis[3].set_title('Demodulated Signal')
dmx_x, dmx_y = make_rect_like(dmx)
axis[3].plot(dmx_x, dmx_y)

axis[4].set_title('Filtered Signal')
axis[4].plot(low_pass_filter(dmx, 10))
```

```python
import wave

with wave.open('test2.wav', 'rb') as wf:
    sample_rate = wf.getframerate()
    num_channels = wf.getnchannels()
    sampwidth = wf.getsampwidth()
    num_frames = wf.getnframes()
    
    signal = np.frombuffer(wf.readframes(num_frames), dtype=np.int16)

max_int16 = 2**15
signal = signal.astype(np.float32) / max_int16

plt.figure(figsize=(15, 3))
plt.plot(signal)
plt.title('Voice Signal')
```

```python
modulated_signal, _ = delta_modulation(signal, 0.005)
print(type(modulated_signal))
```

```python
demouldated_signal = delta_demodulation(mx= modulated_signal, delta= 0.01)
```

```python
figure, axis = plt.subplots(nrows=3, ncols=1)
figure.set_figwidth(10)
figure.set_figheight(10)
figure.tight_layout(h_pad=2)

axis[0].set_title('Signal')
axis[0].plot(signal)

x, y = make_rect_like(modulated_signal[:100])

axis[1].set_title('Modulated Signal')
axis[1].plot(x, y)

axis[2].set_title('Demodulated Signal')
axis[2].plot(demouldated_signal)
```

```python
def export(pcm_float:np.ndarray, samp_width:int, framerate:int, channeles:int, path:str):
    pcm_int16 = (pcm_float * 2**15).astype(np.int16)
    with wave.open(path, 'wb') as wf:
        wf.setnchannels(channeles)
        wf.setsampwidth(samp_width)
        wf.setframerate(framerate)
        wf.writeframes(pcm_int16.tobytes())
```

```python
export(demouldated_signal, sampwidth, sample_rate, num_channels, 'output2.wav')
```
